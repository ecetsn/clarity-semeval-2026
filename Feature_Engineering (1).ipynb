**Stopword listeleri + tokenize**

```python
import re
from sklearn.feature_extraction.text import ENGLISH_STOP_WORDS

STOP_SK = set(ENGLISH_STOP_WORDS)

def tokenize(text: str):
    """Lowercase + basit kelime tokenizasyonu."""
    if not isinstance(text, str):
        return []
    text = text.lower()
    return re.findall(r"[a-z']+", text)

def stopword_ratio(tokens, stop_set=STOP_SK):
    """tokens içindeki stopword oranı."""
    if not tokens:
        return 0.0
    stop_count = sum(1 for t in tokens if t in stop_set)
    return stop_count / len(tokens)
```

```python
def compute_stopword_features_row(row):
    q_text = row.get("interview_question", "")
    a_text = row.get("interview_answer", "")

    q_tokens = tokenize(q_text)
    a_tokens = tokenize(a_text)

    q_ratio = stopword_ratio(q_tokens)
    a_ratio = stopword_ratio(a_tokens)
    diff   = a_ratio - q_ratio  # cevap soruya göre ne kadar daha stopword'lü?

    return pd.Series({
        "q_stop_ratio": q_ratio,
        "a_stop_ratio": a_ratio,
        "stop_ratio_diff": diff,
    })
```

```python
train_with_stop = train_df_clean.copy()

train_with_stop[["q_stop_ratio", "a_stop_ratio", "stop_ratio_diff"]] = (
    train_with_stop.apply(compute_stopword_features_row, axis=1)
)

# Soru ve cevap token listeleri
train_with_stop["q_tokens"] = train_with_stop["interview_question"].apply(tokenize)
train_with_stop["a_tokens"] = train_with_stop["interview_answer"].apply(tokenize)

# Hangi tokenların stopword olduğunu listeleyen kolonlar
train_with_stop["q_stop_tokens"] = train_with_stop["q_tokens"].apply(
    lambda toks: [t for t in toks if t in STOP_SK]
)
train_with_stop["a_stop_tokens"] = train_with_stop["a_tokens"].apply(
    lambda toks: [t for t in toks if t in STOP_SK]
)


cols_view = [

    "interview_question",
    "interview_answer",
    "q_tokens",
    "a_tokens",
    "q_stop_tokens",
    "q_stop_ratio",
    "a_stop_tokens",
    "a_stop_ratio",
    "stop_ratio_diff",
    "clarity_label",
    "evasion_label",
]

train_with_stop[cols_view].head()
```

```python
import re
import pandas as pd
from sklearn.feature_extraction.text import ENGLISH_STOP_WORDS

# ---------- 1) Basic setup ----------
STOP_SK = set(ENGLISH_STOP_WORDS)

NEGATIONS = {
    "no", "not", "never", "none", "nobody", "nothing", "nowhere",
    "neither", "nor", "cannot", "cant", "can't",
    "dont", "don't", "doesnt", "doesn't", "didnt", "didn't",
    "isnt", "isn't", "arent", "aren't", "wasnt", "wasn't",
    "werent", "weren't", "won't", "wouldnt", "wouldn't",
    "shouldnt", "shouldn't", "couldnt", "couldn't"
}

HEDGES = {
    "maybe", "perhaps", "probably", "possibly",
    "apparently", "seem", "seems", "seemed",
    "appear", "appears", "appeared",
    "roughly", "around", "about",
    "sort", "kinda", "kind", "somewhat",
    "guess", "think", "believe", "suggest", "suggests",
    "likely", "unlikely"
}

def tokenize(text: str):
    if not isinstance(text, str):
        return []
    text = text.lower()
    return re.findall(r"[a-z']+", text)

def stopword_ratio(tokens, stop_set=STOP_SK):
    if not tokens:
        return 0.0
    stop_count = sum(1 for t in tokens if t in stop_set)
    return stop_count / len(tokens)

def count_and_ratio(tokens, vocab: set):
    if not tokens:
        return 0, 0.0
    cnt = sum(1 for t in tokens if t in vocab)
    return cnt, cnt / len(tokens)


# ---------- 2) Build feature dataframe from RAW train ----------
train_feats = train_raw.copy()

# tokens
train_feats["q_tokens"] = train_feats["interview_question"].apply(tokenize)
train_feats["a_tokens"] = train_feats["interview_answer"].apply(tokenize)

# stopword features
train_feats["q_stop_tokens"] = train_feats["q_tokens"].apply(
    lambda toks: [t for t in toks if t in STOP_SK]
)
train_feats["a_stop_tokens"] = train_feats["a_tokens"].apply(
    lambda toks: [t for t in toks if t in STOP_SK]
)

# count + ratio
train_feats["q_stop_count"] = train_feats["q_stop_tokens"].apply(len)
train_feats["a_stop_count"] = train_feats["a_stop_tokens"].apply(len)

train_feats["q_stop_ratio"] = train_feats["q_tokens"].apply(stopword_ratio)
train_feats["a_stop_ratio"] = train_feats["a_tokens"].apply(stopword_ratio)
train_feats["stop_ratio_diff"] = train_feats["a_stop_ratio"] - train_feats["q_stop_ratio"]

# negation features
q_neg = train_feats["q_tokens"].apply(lambda toks: count_and_ratio(toks, NEGATIONS))
a_neg = train_feats["a_tokens"].apply(lambda toks: count_and_ratio(toks, NEGATIONS))

train_feats["q_neg_count"]  = q_neg.apply(lambda x: x[0])
train_feats["q_neg_ratio"]  = q_neg.apply(lambda x: x[1])
train_feats["a_neg_count"]  = a_neg.apply(lambda x: x[0])
train_feats["a_neg_ratio"]  = a_neg.apply(lambda x: x[1])
train_feats["neg_ratio_diff"] = train_feats["a_neg_ratio"] - train_feats["q_neg_ratio"]

# hedge features
q_hedge = train_feats["q_tokens"].apply(lambda toks: count_and_ratio(toks, HEDGES))
a_hedge = train_feats["a_tokens"].apply(lambda toks: count_and_ratio(toks, HEDGES))

train_feats["q_hedge_count"] = q_hedge.apply(lambda x: x[0])
train_feats["q_hedge_ratio"] = q_hedge.apply(lambda x: x[1])
train_feats["a_hedge_count"] = a_hedge.apply(lambda x: x[0])
train_feats["a_hedge_ratio"] = a_hedge.apply(lambda x: x[1])
train_feats["hedge_ratio_diff"] = train_feats["a_hedge_ratio"] - train_feats["q_hedge_ratio"]

# ---------- 3) Quick preview ----------
cols_view = [
    "interview_question",
    "interview_answer",
    "clarity_label",
    "evasion_label",
    "q_stop_count", "q_stop_ratio",
    "a_stop_count", "a_stop_ratio", "stop_ratio_diff",
    "q_neg_count", "q_neg_ratio",
    "a_neg_count", "a_neg_ratio", "neg_ratio_diff",
    "q_hedge_count", "q_hedge_ratio",
    "a_hedge_count", "a_hedge_ratio", "hedge_ratio_diff",
]

print("train_feats shape:", train_feats.shape)
display(train_feats[cols_view].head(5))
```

```python
# ============================
# Pragmatic Features + Save + Loader (TRAIN/TEST only)
# Requires: train_df_clean, test_df_clean (may be None)
# ============================

import re
from pathlib import Path
import pandas as pd
from sklearn.feature_extraction.text import ENGLISH_STOP_WORDS

# 0) Use your existing cleaned tables (single source of truth)
train_df = train_df_clean.copy()
test_df  = test_df_clean.copy() if test_df_clean is not None else None

QCOL = "interview_question"
ACOL = "interview_answer"

STOP_SK = set(ENGLISH_STOP_WORDS)

NEGATIONS = {
    "no","not","never","none","nobody","nothing","nowhere","neither","nor",
    "cannot","cant","can't",
    "dont","don't","doesnt","doesn't","didnt","didn't",
    "isnt","isn't","arent","aren't","wasnt","wasn't","werent","weren't",
    "won't","wouldnt","wouldn't","shouldnt","shouldn't","couldnt","couldn't"
}

HEDGES = {
    "maybe","perhaps","probably","possibly",
    "apparently","seem","seems","seemed","appear","appears","appeared",
    "roughly","around","about",
    "sort","kinda","kind","somewhat",
    "guess","think","believe","suggest","suggests",
    "likely","unlikely"
}

HEDGE_PHRASES = [r"\bsort of\b", r"\bkind of\b", r"\bi think\b"]
TOKEN_RE = re.compile(r"[a-z']+")

def tokenize(text: str):
    if not isinstance(text, str):
        return []
    text = text.lower().strip()
    if not text:
        return []
    return TOKEN_RE.findall(text)

def stopword_ratio(tokens):
    if not tokens:
        return 0.0
    return sum(1 for t in tokens if t in STOP_SK) / len(tokens)

def count_and_ratio(tokens, vocab: set):
    if not tokens:
        return 0, 0.0
    cnt = sum(1 for t in tokens if t in vocab)
    return cnt, cnt / len(tokens)

def phrase_count(text: str, patterns):
    if not isinstance(text, str):
        return 0
    t = text.lower()
    return sum(len(re.findall(p, t)) for p in patterns)

def build_pragmatic_features(df: pd.DataFrame) -> pd.DataFrame:
    feats = pd.DataFrame(index=df.index)

    q_tokens = df[QCOL].apply(tokenize)
    a_tokens = df[ACOL].apply(tokenize)

    feats["q_token_len"] = q_tokens.apply(len)
    feats["a_token_len"] = a_tokens.apply(len)

    feats["q_stop_ratio"] = q_tokens.apply(stopword_ratio)
    feats["a_stop_ratio"] = a_tokens.apply(stopword_ratio)
    feats["stop_ratio_diff"] = feats["a_stop_ratio"] - feats["q_stop_ratio"]

    q_neg = q_tokens.apply(lambda toks: count_and_ratio(toks, NEGATIONS))
    a_neg = a_tokens.apply(lambda toks: count_and_ratio(toks, NEGATIONS))
    feats["q_neg_count"] = q_neg.apply(lambda x: x[0])
    feats["q_neg_ratio"] = q_neg.apply(lambda x: x[1])
    feats["a_neg_count"] = a_neg.apply(lambda x: x[0])
    feats["a_neg_ratio"] = a_neg.apply(lambda x: x[1])
    feats["neg_ratio_diff"] = feats["a_neg_ratio"] - feats["q_neg_ratio"]

    q_hedge = q_tokens.apply(lambda toks: count_and_ratio(toks, HEDGES))
    a_hedge = a_tokens.apply(lambda toks: count_and_ratio(toks, HEDGES))
    feats["q_hedge_count"] = q_hedge.apply(lambda x: x[0])
    feats["q_hedge_ratio"] = q_hedge.apply(lambda x: x[1])
    feats["a_hedge_count"] = a_hedge.apply(lambda x: x[0])
    feats["a_hedge_ratio"] = a_hedge.apply(lambda x: x[1])
    feats["hedge_ratio_diff"] = feats["a_hedge_ratio"] - feats["q_hedge_ratio"]

    feats["q_hedge_phrase_count"] = df[QCOL].apply(lambda x: phrase_count(x, HEDGE_PHRASES))
    feats["a_hedge_phrase_count"] = df[ACOL].apply(lambda x: phrase_count(x, HEDGE_PHRASES))

    return feats

def sanity_check(feats: pd.DataFrame, name: str):
    print(f"\n=== Sanity: {name} ===")
    print("rows:", len(feats), "cols:", feats.shape[1])
    print("total NaN:", int(feats.isna().sum().sum()))
    print("q_token_len==0:", int((feats["q_token_len"] == 0).sum()))
    print("a_token_len==0:", int((feats["a_token_len"] == 0).sum()))

    ratio_cols = ["q_stop_ratio","a_stop_ratio","q_neg_ratio","a_neg_ratio","q_hedge_ratio","a_hedge_ratio"]
    bad_ratio = {c: int(((feats[c] < 0) | (feats[c] > 1)).sum()) for c in ratio_cols}
    bad_ratio = {k:v for k,v in bad_ratio.items() if v > 0}
    print("bad ratios (should be empty):", bad_ratio if bad_ratio else "{}")

# 1) Build features
train_feats = build_pragmatic_features(train_df)
sanity_check(train_feats, "train")
train_enriched = pd.concat([train_df.reset_index(drop=True), train_feats.reset_index(drop=True)], axis=1)

test_enriched = None
if test_df is not None:
    test_feats = build_pragmatic_features(test_df)
    sanity_check(test_feats, "test")
    test_enriched = pd.concat([test_df.reset_index(drop=True), test_feats.reset_index(drop=True)], axis=1)

# 2) Save (avoid OUT_DIR name clash; use FEAT_DIR)
FEAT_DIR = Path("./artifacts_features")
FEAT_DIR.mkdir(parents=True, exist_ok=True)

train_enriched.to_parquet(FEAT_DIR / "train_enriched.parquet", index=False)
if test_enriched is not None:
    test_enriched.to_parquet(FEAT_DIR / "test_enriched.parquet", index=False)

# Save feature column list for others
pd.Series(list(train_feats.columns)).to_csv(FEAT_DIR / "pragmatic_feature_columns.csv", index=False, header=False)

print("\nSaved in:", FEAT_DIR.resolve())
print("Files:", [p.name for p in FEAT_DIR.glob("*")])

# 3) Loader for Ece/Ihsan
class PragmaticFeatureLoader:
    def __init__(self, base_dir="./artifacts_features", label_col="clarity_label"):
        self.base_dir = Path(base_dir)
        self.label_col = label_col

    @staticmethod
    def feature_columns():
        return [
            "q_token_len","a_token_len",
            "q_stop_ratio","a_stop_ratio","stop_ratio_diff",
            "q_neg_count","q_neg_ratio","a_neg_count","a_neg_ratio","neg_ratio_diff",
            "q_hedge_count","q_hedge_ratio","a_hedge_count","a_hedge_ratio","hedge_ratio_diff",
            "q_hedge_phrase_count","a_hedge_phrase_count",
        ]

    def load(self, split: str) -> pd.DataFrame:
        return pd.read_parquet(self.base_dir / f"{split}_enriched.parquet")

    def get_xy(self, split: str, qcol=QCOL, acol=ACOL):
        df = self.load(split)
        text = (df[qcol].fillna("") + " " + df[acol].fillna("")).str.strip()
        X = df[self.feature_columns()].copy()
        y = df[self.label_col] if self.label_col in df.columns else None
        return text, X, y

loader = PragmaticFeatureLoader(str(FEAT_DIR))
print("\nLoader check:", loader.feature_columns())
```

```python
import pandas as pd

def _empty_answer_count(df, acol="interview_answer"):
    if df is None or acol not in df.columns:
        return None
    return int(df[acol].fillna("").astype(str).str.strip().eq("").sum())

def _rows(df):
    return None if df is None else int(df.shape[0])

# 1) Detect which objects exist
candidates = {}
for name in ["train_raw", "train_df_clean", "train_df", "train_feats", "train_enriched"]:
    if name in globals():
        obj = globals()[name]
        if isinstance(obj, pd.DataFrame):
            candidates[name] = obj

print("=== Inventory (rows, empty answers) ===")
for name, df_ in candidates.items():
    print(f"{name:15s} rows={_rows(df_)}  empty_answer={_empty_answer_count(df_)}")

# 2) Decide the "source of truth" for your pipeline
# Prefer train_enriched if you already saved/used it; else train_feats; else train_df_clean; else train_raw.
priority = ["train_enriched", "train_feats", "train_df_clean", "train_raw", "train_df"]
source_name = next((n for n in priority if n in candidates), None)

if source_name is None:
    raise RuntimeError("No train dataframe found among: train_raw/train_df_clean/train_df/train_feats/train_enriched")

src = candidates[source_name]
before_rows = _rows(src)
before_empty = _empty_answer_count(src)

print(f"\n=== Using SOURCE = {source_name} ===")
print("Before drop -> rows:", before_rows, " empty_answer:", before_empty)

# 3) If there is at least 1 empty answer, drop it HERE (on the correct table)
if before_empty and before_empty > 0:
    mask = src["interview_answer"].fillna("").astype(str).str.strip().ne("")
    src2 = src.loc[mask].reset_index(drop=True)

    after_rows = _rows(src2)
    after_empty = _empty_answer_count(src2)

    print("\n=== After drop ===")
    print("After drop -> rows:", after_rows, " empty_answer:", after_empty)
    print("Dropped rows:", before_rows - after_rows)

    # 4) Put back to the same variable name (so you don't get confused later)
    globals()[source_name] = src2

    # 5) If your pipeline uses build_pragmatic_features, rebuild features + enriched from the DROPPED version
    if "build_pragmatic_features" in globals():
        # Determine which base df to enrich:
        base_df = src2

        # Rebuild features and enriched
        new_feats = build_pragmatic_features(base_df)
        new_enriched = pd.concat([base_df.reset_index(drop=True), new_feats.reset_index(drop=True)], axis=1)

        # Overwrite the typical names so downstream code is consistent
        train_feats = new_feats
        train_enriched = new_enriched

        print("\nRebuilt train_feats/train_enriched from dropped data.")
        print("train_feats shape:", train_feats.shape)
        print("train_enriched shape:", train_enriched.shape)

        # Resave if FEAT_DIR exists
        if "FEAT_DIR" in globals():
            train_enriched.to_parquet(FEAT_DIR / "train_enriched.parquet", index=False)
            print("Resaved:", (FEAT_DIR / "train_enriched.parquet"))
else:
    print("\nNo empty answers in SOURCE. Nothing to drop.")
```
